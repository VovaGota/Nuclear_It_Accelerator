<h2>Построение эллипса по его параметрам</h2>

Пусть нам дан эллипс, заданный своей квадратичной формой
$$
\gamma x^2 + 2 \alpha xy + \beta y^2=\varepsilon.
$$
Эту выражение более удобно записать в матричном виде
$$
\mathbf{X}^T
\Phi
\mathbf{X}
=\varepsilon,
\hspace{1cm}
\mathbf{X}=
\begin{pmatrix}
x \\
y
\end{pmatrix},
\hspace{1cm}
\Phi=
\begin{pmatrix}
\gamma & \alpha \\
\alpha & \beta
\end{pmatrix}.
$$
Определитель матрицы квадратичной формы $\Phi$ положим равным единице, т.е. примем условие нормировки
$$
\gamma\beta-\alpha^2=1,
$$
тогда площадь эллипса будет равна $\pi\varepsilon$.

Рассмотрим следующий интересный вопрос: каким наилучшим образом нарисовать этот эллипс?


```python
import numpy as np

import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
opts.defaults(opts.Curve(line_width=2))

# Для примера, дальше будем работать с такими коэффициентами:
eps = 5
bet = 4
alf = 2.5
gam = (1+alf**2)/bet

Fi = np.array([[gam,alf],[alf,bet]])
print("Матрица квадратичной формы:")
print(Fi)

print("\nМаксимальные координаты эллипса по осям равны:")
print("x_max = ", np.sqrt(eps*bet))
print("x'_max = ", np.sqrt(eps*gam))
```

<h3>Способ 0. Прямолинейный</h3>

Для начала построим заданный эллипс достаточно прямолинейным путём. Очевидно, что каждому заданному $x$ соответствуют две точки на эллипсе: $y_{min}$ и $y_{max}$. Найдём эти точки, для чего рассмотрим уравнение эллипса как квадратное уравнение и решим его:
$$
y_{min,max}=\frac{-\alpha x \pm \sqrt{\varepsilon\beta - x^2}}{\beta}.
$$

Сейчас можно просто пройти по $x$ от $-\sqrt{\varepsilon\beta}$ до $+\sqrt{\varepsilon\beta}$ и получить две ветви эллипса - верхнюю и нижнюю.

```python
x_ell = np.linspace(-np.sqrt(eps*bet)+0.001,np.sqrt(eps*bet)-0.001,100)
Detr = bet*eps-x_ell**2
y1_ell = (-(alf*x_ell) + np.sqrt(Detr))/bet
y2_ell = (-(alf*x_ell) - np.sqrt(Detr))/bet

branch1 = hv.Curve((x_ell,y1_ell)).redim(x=hv.Dimension("X", range=(-6, 6)), y=hv.Dimension("Y", range=(-6, 6))).opts(width=350, show_grid=True)
branch2 = hv.Curve((x_ell,y2_ell)).redim(x=hv.Dimension("X", range=(-6, 6)), y=hv.Dimension("Y", range=(-6, 6))).opts(width=350, show_grid=True)
branch1*branch2
```

<h3>Способ 1. Использование спектрального разложения матрицы</h3>

Можно к решению поставленной задачи подойти другим путём: мы можем легко расставить точки на окружности; также мы знаем, что линейное преобразование отображает окружность в эллипс. Следовательно, чтобы расставить точки на эллипсе, нам нужно найти такое линейное отображение, которое переводит окружность в заданный эллипс. Оказывается, можно построить два таких отображения!

Начнём со способа, использующего <a href="https://ru.wikipedia.org/wiki/%D0%A1%D0%BF%D0%B5%D0%BA%D1%82%D1%80%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D0%B7%D0%BB%D0%BE%D0%B6%D0%B5%D0%BD%D0%B8%D0%B5_%D0%BC%D0%B0%D1%82%D1%80%D0%B8%D1%86%D1%8B">спектральное разложение матрицы</a>. Рассчитаем собственные числа матрицы квадратичной формы $\Phi$:

$$
\det\left(\Phi-\lambda{E}\right)=
\begin{vmatrix}
\gamma-\lambda & \alpha \\
\alpha & \beta-\lambda
\end{vmatrix}=0,
$$

$$
\lambda^2-2\frac{\gamma+\beta}{2}\lambda+1=0,
\hspace{1cm}
\lambda_{1,2}=\frac{\gamma+\beta}{2}\mp\sqrt{\left(\frac{\gamma+\beta}{2}\right)^2-1}.
$$


Используя собственные вектора этой матрицы как столбцы, составим следующую матрицу (здесь мы использовали уравнение на собственные вектора, проистекающие из первой сторки матрицы $\Phi-\lambda{E}$)

$$
P=
\begin{pmatrix}
1                                & 1 \\
-\frac{\gamma-\lambda_1}{\alpha} & -\frac{\gamma-\lambda_2}{\alpha}
\end{pmatrix}.
$$

Отнормировав столбцы этой матрицы, получим ортонормированную матрицу $P_0$. При помощи этой матрицы диагонализируется матрица $\Phi$:

$$
\Phi=P_0 \Lambda P_0^{-1},
\hspace{1cm}
\Lambda=
\begin{pmatrix}
\lambda_1 & 0 \\
0 & \lambda_2
\end{pmatrix}.
$$

Заметим, что для получения данного разложения столбцы матрицы $P$ можно и не ортонормировать, но далее нам потребуется именно ортогональность матрицы, входящей в это разложение. Попутно заметим, что в случае ортогональности матрицы $P_0$ приходим к выводу, что матрицы $\Phi$ и $\Lambda$ не только <a href="https://en.wikipedia.org/wiki/Matrix_similarity">подобны</a>, но и <a href="https://en.wikipedia.org/wiki/Matrix_congruence">конгруэнтны</a>!

<h4>Проверка спектрального разложения матрицы</h4>

Получим собственые числа заданной матрицы, построим матрицу $P_0$ и проверим ее ортогональность, проверим диагонализацию. В пакете linalg есть функция для нахождения собственных чисел и векторов квадратных матриц, проверим и её

```python
lam1 = (gam+bet)/2-np.sqrt(((gam+bet)/2)**2-1)
lam2 = (gam+bet)/2+np.sqrt(((gam+bet)/2)**2-1)
print(F"Собственные числа матрицы квадратичной формы:\nlam1 = {lam1}\nlam2 = {lam2}")

v1 = -(gam-lam1)/alf
v2 = -(gam-lam2)/alf
val1 = np.sqrt(1+v1**2)
val2 = np.sqrt(1+v2**2)
P0 = np.array([[1/val1,1/val2],[v1/val1,v2/val2]])
print("\nМатрица P0:")
print(P0)

print("\nПроверка ортогональности P0. Должна получиться единичная матрица:")
print(np.matmul(P0,np.matrix.transpose(P0)))

Lam = np.array([[lam1,0],[0,lam2]])
P0m1 = np.linalg.inv(P0)
print("\nПроверка разложения. Должна получиться исходная матрица квадратичной формы:")
print(np.matmul(np.matmul(P0,Lam),P0m1))

w, v = np.linalg.eig(Fi)
print ("\nСобственные числа, получаемые при помощи функции eig:\n", w)
print ("Собственные векторы:\n", v)
```

<h4>Построение отображения по способу 1</h4>

Подставим получившееся разложение в уравнение эллипса

$$
\mathbf{X}^T
P_0 \Lambda P_0^{-1}
\mathbf{X}
=\varepsilon.
$$

Поскольку $P_0$ - ортогональная матрица, то $P_0=\left(P_0^{-1}\right)^{-1}=\left(P_0^T\right)^{-1}=\left(P_0^{-1}\right)^T$, и это уравнение можно записать в слеующем виде

$$
(P_0^{-1}\mathbf{X})^T
\Lambda 
(P_0^{-1}\mathbf{X})
=\varepsilon.
$$

Таким образом, в повёрнутых координатах $\mathbf{\widetilde{X}}$, связанных с начальными координатами $\mathbf{X}$ следующим образом

$$
\mathbf{X} = P_0 \mathbf{\widetilde{X}},
$$

уравнение эллипса представляется в каноническом виде

$$
\mathbf{\widetilde{X}}^T
\Lambda
\mathbf{\widetilde{X}}
=\varepsilon.
$$

Отсюда сразу получаем полуоси эллипса:

$$
a_{1,2}=\sqrt\frac{\varepsilon}{\lambda_{1,2}}.
$$

В повёрнутых координатах эллипс можно обойти следующим образом: 

$$
\widetilde{x} = a_1\cos{\theta},
\hspace{1cm}
\widetilde{y} = a_2\sin{\theta},
\hspace{1cm}
\theta=0...2\pi.
$$

Очевидно, что это можно представить как

$$
\mathbf{\widetilde{X}}=
\begin{pmatrix}
\sqrt\frac{\varepsilon}{\lambda_1} & 0 \\
0 & \sqrt\frac{\varepsilon}{\lambda_2}
\end{pmatrix}
\begin{pmatrix}
\cos{\theta} \\
\sin{\theta}
\end{pmatrix}.
$$

Окончательно получаем отображение единичной окружности на заданный эллипс:

$$
\mathbf{X}=
M_1
\begin{pmatrix}
\cos{\theta} \\
\sin{\theta}
\end{pmatrix},
\hspace{1cm}
M_1=
P_0
\begin{pmatrix}
\sqrt\frac{\varepsilon}{\lambda_1} & 0 \\
0 & \sqrt\frac{\varepsilon}{\lambda_2}
\end{pmatrix}
.
$$

Можно рассматривать данное преобразование как последовательность сжатия/растяжения окружности и последующий поворот.

Отметим, что при данном выборе знаков в $\lambda_{1,2}$ оказывается, что $\lambda_1 < \lambda_2$ (напомним, что след положительно определённой матрицы тоже положителен). Видно, что тогда при данной записи точка $\mathbf{\widetilde{X}}$ при $\theta=0$ отображается в точку на большей полуоси эллипса, а при $\theta=\pi/2$ - в точку на меньшей полуоси.

<h4>Проверка способа 1</h4>

Расставим точки на окружности (для демонстрации особенностей методов возьмем три четверти окружности):

```python
ang = np.linspace(0,1.5*np.pi,50)

X = [[np.cos(a),np.sin(a)] for a in ang]

lins1 = np.linspace(0,1,20)
Z1= [[1,y] for y in lins1]

lins2 = np.linspace(-1,1,20)
Z2= [[x,1] for x in lins2]

S0 = hv.Scatter((X)).redim(x=hv.Dimension("X", range=(-1.5, 1.5)), y=hv.Dimension("Y", range=(-1.5, 1.5))).opts(width=350, show_grid=True)
S1 = hv.Scatter((Z1)).redim(x=hv.Dimension("X", range=(-1.5, 1.5)), y=hv.Dimension("Y", range=(-1.5, 1.5))).opts(width=350, show_grid=True)
S2 = hv.Scatter((Z2)).redim(x=hv.Dimension("X", range=(-1.5, 1.5)), y=hv.Dimension("Y", range=(-1.5, 1.5))).opts(width=350, show_grid=True)

S0*S1*S2
```

Строим эллипс согласно первому способу:

```python
LAMM1 = np.array([[np.sqrt(eps/lam1),0],[0,np.sqrt(eps/lam2)]])
M1 = np.matmul(P0,LAMM1)

Y1 = [[np.matmul(M1,X[k])] for k in range(len(X))]
Y11 = [[np.matmul(M1,Z1[k])] for k in range(len(Z1))]
Y12 = [[np.matmul(M1,Z2[k])] for k in range(len(Z2))]

S10 = hv.Scatter((Y1)).redim(x=hv.Dimension("X", range=(-6, 6)), y=hv.Dimension("Y", range=(-6, 6))).opts(width=350, show_grid=True)
S11 = hv.Scatter((Y11)).redim(x=hv.Dimension("X", range=(-6, 6)), y=hv.Dimension("Y", range=(-6, 6))).opts(width=350, show_grid=True)
S12 = hv.Scatter((Y12)).redim(x=hv.Dimension("X", range=(-6, 6)), y=hv.Dimension("Y", range=(-6, 6))).opts(width=350, show_grid=True)

S10*S11*S12
```

<h3>Способ 2. Использование разложения Холецкого</h3>

Второй способ построения отображения окружности на эллипс использует <a href="https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D0%B7%D0%BB%D0%BE%D0%B6%D0%B5%D0%BD%D0%B8%D0%B5_%D0%A5%D0%BE%D0%BB%D0%B5%D1%86%D0%BA%D0%BE%D0%B3%D0%BE">разложение Холецкого</a> - представление симметричной положительно определённой матрицы $A$ в виде $A=LL^{T}$, где $L$ — нижняя треугольная матрица со строго положительными элементами на диагонали. Разложение Холецкого всегда существует и единственно для любой симметричной положительно определённой матрицы. Именно такой является матрица квадратичной формы $\Phi$.

Согласно проведённым в статье по ссылке формулам, в нашем случае нижняя треугольная матрица $L$ равна

$$
L=
\begin{pmatrix}
\sqrt\gamma & 0 \\
\frac{\alpha}{\sqrt\gamma} & \frac{1}{\sqrt\gamma}
\end{pmatrix}.
$$

<h4>Проверка разложения Холецкого</h4>

Вычислим матрицу $L$, с её помощью непосредственно проверим разложение. В пакете linalg есть функция для нахождения матрицы $L$ разложения Холецкого, проверим и её

```python
L = np.array([[np.sqrt(gam),0],[alf/np.sqrt(gam),1/np.sqrt(gam)]])
print("Матрица L, получаемая по формулам:")
print(L)

print("\nМатрица L, получаемая функцией cholesky:")
print(np.linalg.cholesky(Fi))

print("\nПроверка разложения Холецкого. Должна получиться исходная матрица квадратичной формы:")
print(np.matmul(L,np.matrix.transpose(L)))
```

<h4>Построение отображения по способу 2</h4>

Подставим получившееся разложение в уравнение эллипса

$$
\mathbf{X}^T
LL^{T}
\mathbf{X}
=\varepsilon.
$$

Очевидно, это уравнение можно записать следующим образом

$$
\left(L^T\mathbf{X}\right)^T
\left(L^T\mathbf{X}\right)
=\varepsilon.
$$

Таким образом, в координатах $\mathbf{\widetilde{X}}$, связанных с начальными координатами $\mathbf{X}$ следующим образом

$$
\mathbf{X} = \left(L^T\right)^{-1} \mathbf{\widetilde{X}},
$$

уравнение эллипса представляется в каноническом виде

$$
\mathbf{\widetilde{X}}^T
\mathbf{\widetilde{X}}
=\varepsilon.
$$

Как мы увидим из рисунка далее, это отображение сначала сжимает/растягивает окружность, а потом производит горизонтальный линейный сдвиг вдоль оси X (величина сдвига по X зависит от координаты Y, т.е. точки на оси Y не сдвигаются, чем больше координата Y у точки, тем сильней она сдвигается по X, точки с отрицательным Y сдвигаются в другую сторону) - по крайней мере, эта интерпретация имеет место в двумерной плоскости.

Окончательно получаем отображение единичной окружности на заданный эллипс согласно этому способу:

$$
\mathbf{X}=
M_2
\begin{pmatrix}
\cos{\theta} \\
\sin{\theta}
\end{pmatrix},
\hspace{1cm}
M_2=
\left(L^T\right)^{-1}
\begin{pmatrix}
\sqrt\varepsilon & 0 \\
0 & \sqrt\varepsilon
\end{pmatrix} = 
\sqrt\varepsilon \left(L^T\right)^{-1}.
$$

<h4>Проверка способа 2</h4>

```python
M2 = np.sqrt(eps)*np.linalg.inv(np.matrix.transpose(L))

Y2 = [[np.matmul(M2,X[k])] for k in range(len(X))]
Y21 = [[np.matmul(M2,Z1[k])] for k in range(len(Z1))]
Y22 = [[np.matmul(M2,Z2[k])] for k in range(len(Z2))]

S20 = hv.Scatter((Y2)).redim(x=hv.Dimension("X", range=(-6, 6)), y=hv.Dimension("Y", range=(-6, 6))).opts(width=350, show_grid=True)
S21 = hv.Scatter((Y21)).redim(x=hv.Dimension("X", range=(-6, 6)), y=hv.Dimension("Y", range=(-6, 6))).opts(width=350, show_grid=True)
S22 = hv.Scatter((Y22)).redim(x=hv.Dimension("X", range=(-6, 6)), y=hv.Dimension("Y", range=(-6, 6))).opts(width=350, show_grid=True)

S20*S21*S22
```
