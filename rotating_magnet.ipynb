Секторный магнит – дипольный магнит с формой полюсных наконечников в виде сектора круга. В системе секторного магнита траектория осевой заряженной частицы строго перпендикулярна к входной и выходной плоскости магнита, т.е. в криволинейной системе координат
$$
х_{вх}= 0,   х ׳_{вх}= 0,
$$
$$
х_{вых}= 0,  х ׳_{вых}= 0.
$$

Для идеализированного магнита уравнения движения заряженной частицы в горизонтальной и вертикальной плоскостях записываются следующим образом:
$$
x'' − ( k − 1 /𝝆 ) x = -1/𝝆\frac{Δp}{p};
$$
$$
z'' + kz = 0
$$
где ρ – Ларморовский радиус - радиус кривизны траектории частицы в горизонтальной плоскости;

Общая формула для нахождения Ларморовского радиуса:
$$
𝜌 = \frac{p}{eB} = \frac{\betaℽm_0c}{eB}
$$

$$
β = \frac{v}{c};     ℽ = \frac{E}{E_0}
$$

Сокращенная формула, которую будем использовать в работе:
$$
𝜌=\frac{p}{0,03B}
$$
B – магнитная индукция, кгс, p – импульс частицы. ГЭв / с.

В рамках задачи считаем электроны ультрарелятевистскими с p = 0.1 ГЭв/с (β = 0.98)

Для начала построим массив точек, равномерно заполняющих какой-либо эллипс, (пусть он будет ориентирован по главным осям). Вычислим среднеквадратичный эмиттанс данного распределения, сравним его с теоретическим.

```python
import numpy as np
import holoviews as hv
hv.extension('bokeh')

#Размеры начального эллипса и число точек в нём
ax = 5
ay = 2
N = 2500

#Берём и расставляем (с небольшим запасом) точки по прямоугольной области так,
#чтобы в начальный эллипс попало указаное ранее число точек
#Здесь мы расставилит точки случайным образом, хотя прямоугольную область можно было легко заполнить равномерно
N1 = int(N*4/3.14*1.1)
x = np.random.uniform(low=-ax, high=ax, size=N1)
y = np.random.uniform(low=-ay, high=ay, size=N1)

def ini_ell(x,y):
    return (x/ax)**2 + (y/ay)**2

#Из всех точек формируем массив тех точек, которые находятся внутри начального эллипса
#Обрезаем этот массив до указанного размера
P0 = [[x[k],y[k]] for k in range(N1) if ini_ell(x[k],y[k]) <=1]
if len(P0) > N:
    P0=P0[:N]
N=len(P0)

#Выводим точки
dots1 = hv.Points((P0)).opts(title="Ininial ellipse")
dots1.redim(x=hv.Dimension("x", range=(-5.5, 5.5)), y=hv.Dimension("x'", range=(-5.5, 5.5)))
```

```python
#Функция расчёта среднеквадратичного эмиттанса
def calc_eps_rms(P):
    X2 = 0
    XY = 0
    Y2 = 0
    for k in range(N):
        x = P[k][0]
        y = P[k][1]
        X2 += x*x
        XY += x*y
        Y2 += y*y
    rms2 = (X2*Y2 - XY*XY)/(N**2)
    return np.sqrt(rms2)

#Вывод среднеквадратичного эмиттанса начального распределения
print("Inintial eps_rms = ",calc_eps_rms(P0))
print("Theoretical eps = ", ax*ay)
```

***Расчёт параметров эллипса Флоке***

В данном приближении решения  задачи положим, что поле в магните постоянно и однородно  (k = 0) и в пучке отсутствует разброс частиц по импульсу ( ∆ p = 0 ).

Решение уравнений движения в матричном виде запишется как
$$
\begin{pmatrix}
x \\
x' \\
\end{pmatrix}=
\begin{pmatrix}
\cos{𝝋} & ρ sin{𝝋} \\
-1/ρ*sin{𝝋} & cos{𝝋} \\
\end{pmatrix}
\begin{pmatrix}
x_0 \\
x_0' \\
\end{pmatrix}
$$
$$
\begin{pmatrix}
z \\
z' \\
\end{pmatrix}=
\begin{pmatrix}
1 & ρ𝝋 \\
0 & 1 \\
\end{pmatrix}
\begin{pmatrix}
z_0 \\
z_0' \\
\end{pmatrix}
$$
Где 𝝋 - пролетный угол магнита, 𝜌 - ларморовский радиус

Рассмотрим колебания пучка в плоскости xx'.
В данном случае дипольный магнит можно рассматривать как фокусирующую линзу с матрицей $M_{dipole}$ и фокусным расстоянием F
$$
F = 1/M_{dipole21} = \frac{𝜌}{sin{ϕ}}
$$
Расчитаем параметры начального эллипса Флоке. На рисунке U – расстояние от источника частиц до магнита,  V – расстояние от магнита до изображения источника матрицы передачи этих участков. Построим общую матрицу системы U|M|V
Выберем
$$
 U = 2F
$$

 Матрица перехода системы будет равна
$$
M =
\begin{pmatrix}
1 & v \\
0 & 1 \\
\end{pmatrix}
\begin{pmatrix}
\cos{𝝋} & ρ sin{𝝋} \\
-1/ρ*sin{𝝋} & cos{𝝋} \\
\end{pmatrix}
\begin{pmatrix}
1 & u \\
0 & 1 \\
\end{pmatrix}=
\begin{pmatrix}
cos{𝝋} - v/ρ*sin{𝝋} & -uv/ρ*sin{𝝋} + ucos{𝝋} + ρsin{𝝋}+vcos{𝝋} \\
-1/\rho*sin{𝝋} & cos{𝝋} - u/ρ*sin{𝝋} \\
\end{pmatrix}.
$$

Подберем V таким образом, чтобы элемент $M_{21}$ системы стремился к нулю.  В этом случае на расстоянии V от выхода магнита получается изображение источника.
$$
v→\frac{ucos{ϕ} + 𝜌sin{ϕ}}{u/𝜌sin{ϕ}-cos{ϕ}}
$$

Тогда, косинус набега фазы бетатронных колебаний на период будет равен

$$
\cos{\mu}=\frac{1}{2}Tr(M)=cos{𝝋} -\frac{(u+v)sin{𝝋}}{2p}.
$$

С другой стороны, эта матрица выражается через параметры Твисса
$$
M=
\begin{pmatrix}
\cos{\mu}+\alpha\sin{\mu} & \beta\sin{\mu} \\
-\gamma\sin{\mu} & \cos{\mu}-\alpha\sin{\mu} \\
\end{pmatrix}.
$$
Отсюда находим эти параметры
$$
\beta=(-uv/ρ*sin{𝝋} + ucos{𝝋} + ρsin{𝝋}+vcos{𝝋})/\sin{\mu},
$$
$$
\gamma=sin{𝝋}/ρ\sin{\mu},
$$
$$
\alpha=\left(\cos{\mu}-cos{𝝋} + u/ρ*sin{𝝋} \right)/\sin{\mu}.
$$
Эти величины и есть параметры эллипса Флоке. Оставшуюся величину $\varepsilon$ необходимо подобрать так, чтобы эллипс Флоке описывал начальное представление пучка.

```python
#Определение силы линз для заданного мю
B = 0.5 #магнитная индукция
p = 0.1 #Импульс
phi = np.pi/6 #угол раствора магнита
sinphi = np.sin(phi)
cosphi = np.cos(phi)
rho = p/(0.03*B) # ларморовский радиус
focus = rho/sinphi #фокус системы F = 1/M21
u = 2*focus #подобрано u = 2F
v = (u*cosphi + rho*sinphi)/(u/rho*sinphi - cosphi)-0.3
M12 = cosphi*u+cosphi*v+rho*sinphi-u*v/rho*sinphi
cosmu = cosphi - (u+v)*sinphi/(2*rho)
mu = np.arccos(cosmu)
sinmu = np.sin(mu)
""
#Определение коэффициентов эллиса Флоке в начальной точке
bet = (cosphi*u+cosphi*v+rho*sinphi-u*v/rho*sinphi)/sinmu
gam = sinphi/(rho*sinmu)
alf = (cosmu - cosphi + u/rho*sinphi)/sinmu
gambetp = (gam+bet)/2
gambetm = (gam-bet)/2

#Подобранное значение эффективного эмиттанса
eps = 40
```

***Построение эллипса Флоке по его параметрам***

Для <a href="ellipse.html">реализации</a> построения эллипса Флоке воспользуемся спектральным разложением матрицы квадратичной формы.

```python
#Расчет промежуточных величин для построения эллипса Флоке
lam1 = gambetp + np.sqrt(gambetp**2-1)
lam2 = gambetp - np.sqrt(gambetp**2-1)

a1 = np.sqrt(eps/lam1)
a2 = np.sqrt(eps/lam2)

v1 = -(gam - lam1)/alf
v2 = -(gam - lam2)/alf
P0_11 = 1/np.sqrt(1+v1**2)
P0_12 = 1/np.sqrt(1+v2**2)
P0_21 = v1/np.sqrt(1+v1**2)
P0_22 = v2/np.sqrt(1+v2**2)

#Расчет точек вдоль эллипса Флоке
NEllmax = 100
Teta = [2*np.pi*k/NEllmax for k in range(NEllmax+1)]

EllPoints = [(P0_11*a1*np.cos(T) + P0_12*a2*np.sin(T),
              P0_21*a1*np.cos(T) + P0_22*a2*np.sin(T)) for T in Teta]


#Построение графика эллипса Флоке
EllCurve = hv.Curve( (EllPoints), label='Floquet ellipse' ).opts(color="red")
```



Используя указанные матрицы, можно через каждый период восстанавливать $x$ и $x'$ для каждой частицы ансамбля, попутно вычисляя величину среднеквадратичного эмиттанса.

```python
#Задание коэффициента нелинейности и рассчитываемого числа периодов
c = 0.000
#c = 0.03 #№Здесь уже проявляются интересные нелинейности!!!
Nturns = 4*100

#Задание числа шагов для вывода фазового портрета пучка
Nplots = 11
Nt_to_p = np.arange(0,Nturns+1,int(Nturns/(Nplots-1)))

#Задание числа шагов для построения графика зависимоси среднеквадратичного эмиттанса
#от числа пройденных периодов
Nempts = 11
Nt_to_e = np.arange(0,Nturns+1,int(Nturns/(Nempts-1)))

eps_rms=[]       #Список для хранения величин среднеквадратичного эмиттанса
points_dict = {} #Словарь для хранения фазовых портретов

P=P0.copy()      #Инициализируем распределение точек по начальному
#P=P0            #Особенность Python - так мы только делаем ещё одну ссылку на массив, а не делаем новый массив
                 #Если сделаем так, то при запуске тольк данной ячейки повтороно мы как-бы продолжим идти вдоль канала

#Задание матрицы пустого промежутка
U = np.array([[ 1, u],
               [ 0, 1]])

V = np.array([[ 1, v],
               [ 0, 1]])

#Задание матрицы поворотного диполя
MROT = np.array([[ np.cos(phi), rho*np.sin(phi)],
               [ -1/rho*np.sin(phi), np.cos(phi)]])

#Собственно цикл по периодам, внутри него цикл по частицам
for k in range(Nturns+1):
    if k in Nt_to_p:
        points_dict[k] = hv.Points((P))

    if k in Nt_to_e:
        eps_rms.append(calc_eps_rms(P))

    for l in range(N):
        M1 = np.matmul(V, MROT)
        M = np.matmul(M1, U)
        P[l] = np.matmul(M, P[l])

#Построение графиков - фазовых поретретов пучка при различных значениях числа пройденных периодов
phase_map = hv.HoloMap(points_dict, hv.Dimension(('Nturns')))

#Вывод построенных графиков в одном окне
overlay_graphs = (phase_map*EllCurve).opts(width=500, height=500)
overlay_graphs.opts(legend_position='bottom_right')
```

***Возрастание эмиттанса (emittance dilution)***

По результатам растёта построим график возрастания среднеквадратичного эмиттанса с ростом числа пройденных периодов.

```python
#Вывод графика зависимоси среднеквадратичного эмиттанса от числа пройденных периодов,
#сравнение с величиной эффективного эмиттанса
print ("Effective emitance = ", eps)
rms_plot = hv.Curve((Nt_to_e,eps_rms), "Nturns", "Eps_rms").opts(tools=['hover'], show_grid=True)
rms_plot
```
