Сначала проводим инициализацию библиотек

```python
import numpy as np
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
#hv.extension('matplotlib')
```

```python
alfx = -0.5
betx = 5.0
gamx = (1.0 + alfx**2)/betx
epsx = 10

alfy = 0.3
bety = 2.5
gamy = (1.0 + alfy**2)/bety
epsy = 2

x_max  = np.sqrt(epsx*betx)
xp_max = np.sqrt(epsx*gamx)
y_max  = np.sqrt(epsy*bety)
yp_max = np.sqrt(epsy*gamy)

print(F"x_max = {x_max:10.5f}, x'_max = {xp_max:10.5f}")
print(F"y_max = {y_max:10.5f}, y'_max = {yp_max:10.5f}")
```

Вывод Элипса методом Холетского

```python
betxn = betx / epsx
alfxn = alfx / epsx
gamxn = gamx / epsx

betyn = bety / epsy
alfyn = alfy / epsy
gamyn = gamy / epsy

Qx = np.array([[gamxn,alfxn],[alfxn,betxn]])
Lx = np.linalg.cholesky(Qx)
Mx = np.linalg.inv(np.matrix.transpose(Lx))

Qy = np.array([[gamyn,alfyn],[alfyn,betyn]])
Ly = np.linalg.cholesky(Qy)
My = np.linalg.inv(np.matrix.transpose(Ly))

ang = np.linspace(0,2*np.pi,100)
XO = [[np.cos(a),np.sin(a)] for a in ang]

Xx = [np.matmul(Mx,XO[k]) for k in range(len(XO))]
Xy = [np.matmul(My,XO[k]) for k in range(len(XO))]

xxdim=hv.Dimension("x", range=(-10, 10))
xpdim=hv.Dimension("x'", range=(-10, 10))
yydim=hv.Dimension("y", range=(-10, 10))
ypdim=hv.Dimension("y'", range=(-10, 10))

Sx = hv.Curve((Xx)).redim(x=xxdim, y=xpdim).opts(show_grid=True, color="red")
Sy = hv.Curve((Xy)).redim(x=yydim, y=ypdim).opts(show_grid=True, color="red")
Sx+Sy
```

```python
XO4 = np.array([[x1[k],x2[k],x3[k],x4[k]] for k in range(Np)])
X = [np.matmul(M,XO4[k]) for k in range(Np)]
xx = np.array([X[k][0] for k in range(Np)])
xp = np.array([X[k][1] for k in range(Np)])
yy = np.array([X[k][2] for k in range(Np)])
yp = np.array([X[k][3] for k in range(Np)])

SR1 = hv.Scatter((xx,xp)).redim(x=xxdim, y=xpdim)
SR2 = hv.Scatter((yy,yp)).redim(x=yydim, y=ypdim)
SR3 = hv.Scatter((xx,yy)).redim(x=xxdim, y=yydim)

Mxy = np.array([[np.sqrt(epsx*betx),0],[0,np.sqrt(epsy*bety)]])
Xxy = [np.matmul(Mxy,XO[k]) for k in range(len(XO))]
Sxy = hv.Curve((Xxy)).redim(x=xxdim, y=yydim).opts(show_grid=True, color="red")

Sx*SR1+Sy*SR2+Sxy*SR3
```

теперь засовываем этот бред в fodo канал

```python
#Функция расчёта среднеквадратичного эмиттанса

def ini_ell(x,xp):
    return (x/x_max)**2 + (xp/xp_max)**2

P0 = [[xx[k],xp[k]] for k in range(Np) if ini_ell(xx[k],xp[k]) <=1]
if len(P0) > Np:
    P0=P0[:Np]
Np=len(P0)

def calc_eps_rms(P):
    X2 = 0
    XY = 0
    Y2 = 0
    for k in range(Np):
        x = P[k][0]
        y = P[k][1]
        X2 += x*x
        XY += x*y
        Y2 += y*y
    rms2 = (X2*Y2 - XY*XY)/(Np**2)
    return np.sqrt(rms2)

#Вывод среднеквадратичного эмиттанса начального распределения
print("Inintial eps_rms = ",calc_eps_rms(P0))
print("Theoretical eps = ", x_max*y_max)
```

Параметры эллипса Флоке

```python
#Определение силы линз для заданного мю
mu = np.pi/4
d = 1.5

cosmu = np.cos(mu)
sinmu = np.sin(mu)
f = d/np.sqrt(2*(1-cosmu))
""
#Определение коэффициентов эллиса Флоке в начальной точке
bet = 2*f*np.sqrt((2*f+d)/(2*f-d))
gam = d/(f*f*sinmu)
alf = (cosmu - 1- d/f)/sinmu
gambetp = (gam+bet)/2
gambetm = (gam-bet)/2

#Подобранное значение эффективного эмиттанса
eps = 34
print(f)
```

Построение элипса *Флоке*

```python
#Расчет промежуточных величин для построения эллипса Флоке
lam1 = gambetp + np.sqrt(gambetp**2-1)
lam2 = gambetp - np.sqrt(gambetp**2-1)

a1 = np.sqrt(eps/lam1)
a2 = np.sqrt(eps/lam2)

v1 = -(gam - lam1)/alf
v2 = -(gam - lam2)/alf
P0_11 = 1/np.sqrt(1+v1**2)
P0_12 = 1/np.sqrt(1+v2**2)
P0_21 = v1/np.sqrt(1+v1**2)
P0_22 = v2/np.sqrt(1+v2**2)

#Расчет точек вдоль эллипса Флоке
NEllmax = 100
Teta = [2*np.pi*k/NEllmax for k in range(NEllmax+1)]

EllPoints = [(P0_11*a1*np.cos(T) + P0_12*a2*np.sin(T),
              P0_21*a1*np.cos(T) + P0_22*a2*np.sin(T)) for T in Teta]


#Построение графика эллипса Флоке
EllCurve = hv.Curve( (EllPoints), label='Floquet ellipse' ).opts(color="red")
EllCurve
```

```python
#Задание коэффициента нелинейности и рассчитываемого числа периодов
c = 0.003
#c = 0.03 #№Здесь уже проявляются интересные нелинейности!!!
Nturns = 4*100

#Задание числа шагов для вывода фазового портрета пучка
Nplots = 81
Nt_to_p = np.arange(0,Nturns+1,int(Nturns/(Nplots-1)))

#Задание числа шагов для построения графика зависимоси среднеквадратичного эмиттанса
#от числа пройденных периодов
Nempts = 81
Nt_to_e = np.arange(0,Nturns+1,int(Nturns/(Nempts-1)))

eps_rms=[]       #Список для хранения величин среднеквадратичного эмиттанса
points_dict = {} #Словарь для хранения фазовых портретов

P=P0.copy()      #Инициализируем распределение точек по начальному
#P=P0            #Особенность Python - так мы только делаем ещё одну ссылку на массив, а не делаем новый массив
                 #Если сделаем так, то при запуске тольк данной ячейки повтороно мы как-бы продолжим идти вдоль канала

#Задание матрицы пустого промежутка
M0 = np.array([[ 1, d],
               [ 0, 1]])

#Расчет матрицы тонкой фокусирующей линзы с учётом нелинейности
def MF(x):
    return np.array([[ 1,               0],
                     [ -1/f*(1+c*x**2), 1]])

#Расчет матрицы тонкой дефокусирующей линзы с учётом нелинейности
def MDF(x):
    return np.array([[ 1,               0],
                     [ 1/f*(1+c*x**2), 1]])

#Собственно цикл по периодам, внутри него цикл по частицам
for k in range(Nturns+1):
    if k in Nt_to_p:
        points_dict[k] = hv.Points((P))

    if k in Nt_to_e:
        eps_rms.append(calc_eps_rms(P))

    for l in range(Np):
        M1 = np.matmul(M0, MF(P[l][0]))
        M2 = np.matmul(M0, MDF(P[l][0]))
        M = np.matmul(M2, M1)
        P[l] = np.matmul(M, P[l])

#Построение графиков - фазовых поретретов пучка при различных значениях числа пройденных периодов
phase_map = hv.HoloMap(points_dict, hv.Dimension(('Nturns')))

#Вывод построенных графиков в одном окне
#overlay_graphs = (phase_map * EllCurve).opts(fig_size=200, aspect=1, legend_position='bottom_right')
overlay_graphs = (phase_map*EllCurve).opts(width=500, height=500)
overlay_graphs.opts(legend_position='bottom_right')
#overlay_graphs
```
