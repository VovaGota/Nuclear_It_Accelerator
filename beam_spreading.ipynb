<h2>Расплывание интенсивного пучка в свободном пространстве</h2>

Рассмотрим расплывание интенсивного пучка с током $I$ в свободном пространстве, под действием только собственных полей. Направим ось $z$ вдоль направления движения пучка. Сделаем ряд допущений:

<ul>
    <li>Пусть скорости всех частиц направлены в основном вдоль $z$, приблизительно одинаковы и равны $v$. Другими словами, мы будем работать в параксиальном приближении.</li>
    <li>Пусть эмиттанс пучка равен нулю.</li>
    <li>Пусть в начале своего движения пучок был круглым в сечении, а распределение плотности заряда по сечению было однородным.</li>
    <li>Пусть у всех частиц пучка отсутствует вращение вокруг оси (другими словами, канонический угловой момент равен нулю, частицы стартуют с незамагниченного катода).</li>
    <li>Такой пучок и в дальнйшем будет ламинарным, круглым в сечении, а радиус границы пучка, которой обозначим как $r_b(z)$, всегда совпадает с радиусом крайней частицы.</li>
</ul>

При данных условиях можно показать, что граница пучка описывается следующим дифференциальным уравнением:
$$
\frac{d^2r_b(z)}{dz^2}=\frac{K}{r_b(z)}.
$$
Здесь $K$ - безразмерная величина, которую называют <i>обобщённый первеанс</i>. Эта величина равна
$$
K=\frac{1}{2\pi\varepsilon_0}\frac{qI}{m\gamma^3v^3}=\frac{2}{(\beta\gamma)^3}\frac{I}{I_0},
$$
$I_0=4\pi\varepsilon_0mc^3/q$ - характеристический ток, также называемый в ускорительной литературе током Альфвена. Для электронов $I_0=17.045 \space кА$, для протонов $I_0=31.297 \space МА$.

Для упрощения дальнейшего анализа удобно в этом уравнении перейти к следующим безразмерным переменным:
$$
R=\frac{r_b}{r_0},
\hspace{1cm}
Z=\sqrt{2K}\frac{z}{r_0}.
$$
Здесь $r_0=r_b(0)$ - начальный радиус пучка. В данных переменных уравнение, описывающее расплывание пучка, примет следующий вид:
$$
\frac{d^2R(Z)}{dZ^2}=\frac{1}{2R(Z)}.
$$
Исходя из определения величины $R$, её начальное значение всегда равно единице:
$$
R(0)=1.
$$
Поскольку мы рассматриваем уравнение второго порядка, для его решения важна также начальная производная, т.е. начальный наклон границы пучка. Как видно, наклон границы в размерных и безразмерных переменных связан между собой соотношением
$$
\frac{dR}{dZ}=\frac{1}{\sqrt{2K}}\frac{dr_b}{dz}.
$$

Решим безразмерное уравнение и покажем расплывание пучка вдоль $Z$ при разных начальных её наклонах:

```python
import numpy as np
from scipy.integrate import solve_ivp

import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
opts.defaults(opts.Curve(line_width=1))

#Расплывание огибающей
def Fb(z,R):
    return np.array([R[1], 1/(2.0*R[0])])

z_span = [0, 2.5]
z_eval = np.linspace(z_span[0], z_span[1], 501)

rbp0_arr = np.linspace(0,-2.5,51)

z_cr=[]

def rb(rbp0):
    R0=np.array([1.0, rbp0])
    sol_rb = solve_ivp(Fb, z_span, R0, t_eval=z_eval)
    
    min_ind_rb = np.argmin(sol_rb.y[0])
    z_cr.append(z_eval[min_ind_rb])
    
    return sol_rb.y[0]

beta_dict = {F: hv.Curve((z_eval,rb(F))).redim(x=hv.Dimension("Z"), y=hv.Dimension("R", range=(0, 2.5))) for F in rbp0_arr} 
beta_map = hv.HoloMap(beta_dict, hv.Dimension(('R_prine0'), default=-1))
beta_map.opts(opts.Curve(width=600, show_grid=True))
```

***Расчет оптимального угла наклона***

Изменяя в предыдущей ячейке начальный наклон границы пучка, можно заметить, что при какой-то средней величине начального наклона пучок расплывается медленнее, чем при очень малых и при больших начальных наклонах. Продемонстрируем это, построив график зависимости положения кроссовера пучка (где пучок имеет наименьший размер) от начального наклона.

```python
z_cr_graph = hv.Curve((-rbp0_arr, z_cr)).opts(tools=['hover'], show_grid=True)
z_cr_graph.redim(x=hv.Dimension("-R_prine0"), y=hv.Dimension("Z_cr", range=(0, 1.2)))
```

Видно, что кроссовер максимально удалён от начального положения при $dR/dZ=-0.92413887$, само же максимальное расстояние до кроссовера равно $Z_{cr,max}=2 \cdot 0.54104422=1.08208844$ (см. <a href="https://en.wikipedia.org/wiki/Dawson_function">интеграл Доусона</a>).

***Расплывание протонного пучка***

В качестве примера рассмотрим расплывание протонного пучка из кроссовера. Пусть энергия протонов равна $100 \space кэВ$, ток пучка $0.1 \space А$, начальный радиус $1 \space см$.

Сравним точное решение дифференциального уравнения с приближённым решением в виде степенного ряда, описывающим расплывание из кроссовера:
$$
R(Z)=1+0.25Z^2-0.017Z^3.
$$

```python
I = 0.1              #Ток пучка (в А)
I0 = 31.297*(10**6)  #Характеристический ток протонов (вА)
r0 = 1               #Начальный радиус пучка (в см)
W = 100              #Начальная энергия протонов (в кэВ)
E0 = 938.272*(10**3) #Энергия покоя протонов (в кэВ)

#Вычисление безразмерной границы
R_pb = rb(0)

bg = np.sqrt((W/E0)*(2+W/E0))
K = 2/(bg**3)*I/I0

#Вычисление размерной границы (z в см, r_b в см)
z_pb = (r0/np.sqrt(2*K))*z_eval
r_pb = r0*R_pb

#Вычисление приближённого решения
z_eval_A = np.linspace(z_span[0], z_span[1], 21)
R_pb_A = 1 + 0.25*(z_eval_A**2) - 0.017*(z_eval_A**3)

z_pbA = (r0/np.sqrt(2*K))*z_eval_A
r_pbA = r0*R_pb_A

r_b_protE = hv.Curve((z_pb,  r_pb ), label='Exact solution').redim(x=hv.Dimension("z (cm)"), y=hv.Dimension("r_b (cm)", range=(0, 2.5))).opts(tools=['hover'])
r_b_protA = hv.Points((z_pbA, r_pbA), label='Approximate solution').opts(color="red")

overlay_graphs = (r_b_protE*r_b_protA).opts(width=600, show_grid=True)
overlay_graphs.opts(legend_position='bottom_right')
```

Как видно, при расплывании пучка более чем в два раза точность приближённого решения остаётся очень хорошей!
